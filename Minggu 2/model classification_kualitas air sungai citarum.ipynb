```python
# Import pustaka yang dibutuhkan
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
# Muat dataset
file_path_classification = "C:\\Users\\user\\Documents\\SEMESTER 7\\machine learning\\minggu 2\\Dataset\\CitarumWater.csv"  # Lokasi dataset
data_classification = pd.read_csv(file_path_classification)  # Membaca data menggunakan pandas
```

```python
# Bersihkan kolom 'Temp' (hapus '°C' dan konversi ke float)
data_classification['Temp'] = data_classification['Temp'].str.replace('°C', '').astype(float)  # Menghapus simbol °C dan mengonversi ke angka
```

```python
# Tentukan fitur (X) dan target (y)
X_class = data_classification[['pH', 'Temp', 'TSS', 'DO', 'BOD', 'COD', 'Nitrat', 'FecalColiform', 'Fosfat', 'IP']]  # Fitur (variabel input)
y_class = data_classification['Class']  # Target (variabel output)
```

```python
# Bagi data menjadi data latih dan uji
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_class, y_class, test_size=0.2, random_state=42)  
# 80% data untuk latih, 20% untuk uji, acak berdasarkan seed 42
```

```python
# Standarisasi fitur
scaler = StandardScaler()  # Inisialisasi StandardScaler untuk standarisasi data
X_train_class = scaler.fit_transform(X_train_class)  # Melakukan standarisasi pada data latih
X_test_class = scaler.transform(X_test_class)  # Melakukan standarisasi pada data uji
```

```python
# Model Logistic Regression
logistic_model = LogisticRegression(max_iter=1000)  # Inisialisasi model Logistic Regression dengan iterasi maksimal 1000
logistic_model.fit(X_train_class, y_train_class)  # Melatih model dengan data latih
```

```python
# Prediksi pada data uji
y_pred_class = logistic_model.predict(X_test_class)  # Melakukan prediksi pada data uji
```

```python
# Menghitung metrik evaluasi
accuracy = accuracy_score(y_test_class, y_pred_class)  # Menghitung akurasi
precision = precision_score(y_test_class, y_pred_class, average='weighted')  # Menghitung precision
recall = recall_score(y_test_class, y_pred_class, average='weighted')  # Menghitung recall
f1 = f1_score(y_test_class, y_pred_class, average='weighted')  # Menghitung F1-score
conf_matrix = confusion_matrix(y_test_class, y_pred_class)  # Membuat confusion matrix
```

```python
# Menampilkan Hasil Evaluasi
print(f"Accuracy: {accuracy}")  # Menampilkan akurasi
print(f"Precision: {precision}")  # Menampilkan precision
print(f"Recall: {recall}")  # Menampilkan recall
print(f"F1 Score: {f1}")  # Menampilkan F1-score
print("Confusion Matrix:")  # Menampilkan confusion matrix
print(conf_matrix)  # Menampilkan confusion matrix dalam bentuk matriks
```

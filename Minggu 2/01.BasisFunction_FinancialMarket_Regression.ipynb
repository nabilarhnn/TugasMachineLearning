```python
# Import pustaka yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Muat dataset
file_path = "C:\\Users\\user\\Documents\\SEMESTER 7\\machine learning\\minggu 2\\Dataset\\FinancialMarket.csv"  # Lokasi dataset
data = pd.read_csv(file_path)  # Membaca dataset menggunakan pandas
```

```python
# Extract the x and combined data columns
x = data['x'].values
combined_data = data['combined_data'].values

# Step 3: Visualize the dataset
plt.figure(figsize=(7, 6))

# Plot the combined data
plt.plot(x, combined_data, label='company’s stock price', color='blue')

# Add labels and title
plt.title('Financial Market Response to Earnings Reports')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()  # Ensure there's a label to show in the legend

# Display the plot
plt.show()
```

```python
# Pisahkan fitur (X) dan target (y)
X = data[['x']]  # Fitur (input)
y = data['combined_data']  # Target (output)
```

```python
# Model 1: Vanilla Linear Regression
vanilla_model_full = LinearRegression()  # Inisialisasi model regresi linear
vanilla_model_full.fit(X, y)  # Melatih model menggunakan keseluruhan data
y_pred_vanilla_full = vanilla_model_full.predict(X)  # Prediksi menggunakan model vanilla
```

```python
# Model 2: Radial Basis Function Regression (RBF)
rbf_model_full = KernelRidge(kernel='rbf', gamma=0.5)  # Inisialisasi model RBF dengan kernel Gaussian
rbf_model_full.fit(X, y)  # Melatih model RBF pada data penuh
y_pred_rbf_full = rbf_model_full.predict(X)  # Prediksi menggunakan model RBF
```

```python
# Evaluasi Model Vanilla Linear Regression
mae_vanilla = mean_absolute_error(y, y_pred_vanilla_full)  # Hitung MAE untuk vanilla
mse_vanilla = mean_squared_error(y, y_pred_vanilla_full)  # Hitung MSE untuk vanilla
rmse_vanilla = np.sqrt(mse_vanilla)  # Hitung RMSE untuk vanilla
r2_vanilla = r2_score(y, y_pred_vanilla_full)  # Hitung R² untuk vanilla
```

```python
# Evaluasi Model RBF Regression
mae_rbf = mean_absolute_error(y, y_pred_rbf_full)  # Hitung MAE untuk RBF
mse_rbf = mean_squared_error(y, y_pred_rbf_full)  # Hitung MSE untuk RBF
rmse_rbf = np.sqrt(mse_rbf)  # Hitung RMSE untuk RBF
r2_rbf = r2_score(y, y_pred_rbf_full)  # Hitung R² untuk RBF
```

```python
# Tampilkan Hasil Evaluasi
print("Evaluasi Vanilla Linear Regression:")  # Menampilkan hasil evaluasi untuk vanilla linear regression
print(f"MAE: {mae_vanilla}")  # Tampilkan MAE
print(f"MSE: {mse_vanilla}")  # Tampilkan MSE
print(f"RMSE: {rmse_vanilla}")  # Tampilkan RMSE
print(f"R²: {r2_vanilla}\n")  # Tampilkan R²

print("Evaluasi Radial Basis Function Regression:")  # Menampilkan hasil evaluasi untuk RBF regression
print(f"MAE: {mae_rbf}")  # Tampilkan MAE
print(f"MSE: {mse_rbf}")  # Tampilkan MSE
print(f"RMSE: {rmse_rbf}")  # Tampilkan RMSE
print(f"R²: {r2_rbf}\n")  # Tampilkan R²
```

```python
# Plot hasil prediksi dari kedua model
plt.figure(figsize=(8, 6))  # Mengatur ukuran gambar
plt.plot(X, y, color='blue', label='Data Asli')  # Plot data asli dengan warna biru
plt.plot(X, y_pred_vanilla_full, color='green', linestyle='--', label='Vanilla Regression')  # Plot prediksi vanilla regression dengan garis hijau putus-putus
plt.plot(X, y_pred_rbf_full, color='red', label='Basis Function Regression Model')  # Plot prediksi RBF regression dengan garis merah
plt.title('Vanilla Regression vs Basis Function Regression Model')  # Judul grafik
plt.xlabel('X')  # Label untuk sumbu X
plt.ylabel('Y')  # Label untuk sumbu Y
plt.legend()  # Tampilkan legenda untuk penjelasan
plt.show()  # Tampilkan plot
```
